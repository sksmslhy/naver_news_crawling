```python
cleanbot_cnt = 0

url_link = []
try:
    a = 1
    while True:
        urls = url_pasing(driver)
        url_link.extend(urls)
        scroll_down(driver)
        a+=1
        driver.find_element_by_link_text(f'{a}').click()
        #sp_nws1 > div.news_wrap.api_ani_send > div > div.news_info > div.info_group > a:nth-child(3)
        
except:
    print(url_link)


target_date= []
target_title = []
target_content = []
target_coment = []
target_recomm = []
target_unrecomm = []

for c in range(0, len(url_link)):
    url = url_link[c]
    time.sleep(random.random())
    driver.get(url)

    scroll_down(driver)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 날짜
    try:
        date_time = driver.find_elements_by_css_selector('div.sponsor > span.t11')
        date = date_time[0].text 
    except:
        date = f"가져오지 못했습니다. 주소를 확인하세요.{url}"

    # 제목
    try:
        title = soup.find('div', class_='article_info').find('h3', class_='tts_head').get_text() 
        # #main_content > div.article_header > div.article_info
    except:
        title = f"가져오지 못했습니다. 주소를 확인하세요.{url}"

    # 내용
    try:
        content = soup.find('div', class_='article_body _font_setting_target size3').find('div', class_='_article_body_contents').get_text() # 내용
        content = content.replace("\n","")
        #print(content)
    except:
        content = f"가져오지 못했습니다. 주소를 확인하세요.{url}"

    #댓글
    # 댓글 페이지로 이동
    com_url = url+"&m_view=1"
    driver.get(com_url)
    time.sleep(3)

    # 클린봇 해제 (한 번만 실행하면 됨.)
    if cleanbot_cnt == 0 :
        cleanbot = driver.find_element_by_css_selector('a.u_cbox_cleanbot_setbutton')
        cleanbot.click()
        time.sleep(1)
        cleanbot_disable = driver.find_element_by_xpath("//input[@id='cleanbot_dialog_checkbox_cbox_module']")
        cleanbot_disable.click()
        time.sleep(1)
        driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[4]/button').click()
        time.sleep(3)
        cleanbot_cnt += 1


    #더보기 계속 클릭하기
    while True:
        try:
            btn_more = driver.find_element_by_css_selector('a.u_cbox_btn_more')
            btn_more.click()
            time.sleep(1)
        except:
            break

    scroll_down(driver)
    coment =[]
    try:
        txts = driver.find_elements_by_css_selector('span.u_cbox_contents')
        for t in txts:
            txt = t.text
            coment.append(txt)        
    except:
        print("댓글이 없습니다.")
        txt = '댓글이 없습니다.'
        coment.append(txt)

        # 추천수
    recomms = []
    goods = driver.find_elements_by_css_selector('em.u_cbox_cnt_recomm')
    for g in goods:
        good = g.text
        recomms.append(good)

    # 반대수
    unrecomms = []
    bads = driver.find_elements_by_css_selector('em.u_cbox_cnt_unrecomm')
    for b in bads:
        bad = b.text
        unrecomms.append(bad)

    target_date.append(date)
    target_title.append(title)
    target_content.append(content)
    target_coment.append(coment)
    target_recomm.append(recomms)
    target_unrecomm.append(unrecomms)
```

```python
naver_new = pd.DataFrame()
naver_new['날짜'] = pd.Series(target_date)
naver_new['제목'] = pd.Series(target_title)
naver_new['내용'] = pd.Series(target_content)
naver_new['댓글'] = pd.Series(target_coment)
naver_new['댓글별 추천수'] = pd.Series(target_recomm)
naver_new['댓글별 반대수'] = pd.Series(target_unrecomm)

naver_new.head()
```
