```python
url_link = []
try:
    a = 1
    while True:
        urls = url_pasing(driver)
        url_link.extend(urls)
        scroll_down(driver)
        a+=1
        driver.find_element_by_link_text(f'{a}').click()
        
except:
    print(url_link)


target_date= []
target_title = []
target_content = []
target_coment = []
for c in range(0, len(url_link)):
    url = url_link[c]
    time.sleep(random.random())
    driver.get(url)

    scroll_down(driver)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 날짜
    try:
        date_time = driver.find_elements_by_css_selector('div.sponsor > span.t11')
        date = date_time[0].text 
    except:
        date = f"가져오지 못했습니다. 주소를 확인하세요.{url}"

    # 제목
    try:
        title = soup.find('div', class_='article_info').find('h3', class_='tts_head').get_text() 
    except:
        title = f"가져오지 못했습니다. 주소를 확인하세요.{url}"

    # 내용
    try:
        content = soup.find('div', class_='article_body _font_setting_target size3 font1').find('div', class_='_article_body_contents').get_text() # 내용
        content = re.sub('\s+', '', content)
        content = re.sub('[ㄱ-ㅎㅏ-ㅣ!?~,".\n\r#\ufeff\u200d]', '', content)
        content = re.sub('·', '', content)
        content = re.sub('※', '', content)
        content = re.sub('☞', '', content)
        content.replace('\xa0', " ")
    except:
        content = f"가져오지 못했습니다. 주소를 확인하세요.{url}"

    #댓글
    scroll_down(driver)
    coment =[]
    try:
        txts = driver.find_elements_by_css_selector('span.u_cbox_contents')
        for t in txts:
            txt = t.text
            coment.append(txt)
    except:
        print("댓글이 없습니다.")
        txt = '댓글이 없습니다.'
        coment.append(txt)

    target_date.append(date)
    target_title.append(title)
    target_content.append(content)
    target_coment.append(coment)


naver_new = pd.DataFrame()
naver_new['날짜'] = pd.Series(target_date)
naver_new['제목'] = pd.Series(target_title)
naver_new['내용'] = pd.Series(target_content)
naver_new['댓글'] = pd.Series(target_coment)

#CSV형태로 저장하기
naver_new.to_csv('C:/Users/User/Desktop/test/'+'조선일보_중국(191101-200820).csv', encoding='utf-8-sig', index=False)

#엑셀 형태로 저장하기
#naver_new.to_excel('C:/Users/User/Desktop/test/'+'조선일보_코로나(200501-201231).xls', index=False)
```
